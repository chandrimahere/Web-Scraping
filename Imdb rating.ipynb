```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```python
#send request
url = 'https://www.imdb.com/search/title/?genres=drama&groups=top_250&sort=user_rating,desc'
page=requests.get(url)
page.text
```

```python
soup=BeautifulSoup(page.text,'html.parser')
```

```python
#read index_no
index_no=[]
index_no_elem=soup.find_all(class_='lister-item-index unbold text-primary')

for item in index_no_elem:
    index_no.append(item.text)
```

```python
#read moviename
movie_name=[]
movie_name_elem=soup.find_all(class_='lister-item-header')

for item in movie_name_elem:
    movie_name.append(item.text)
    
movie_name
```

```python
#read year_of_release
year_of_release=[]
year_of_release_elem=soup.find_all(class_='lister-item-year text-muted unbold')

for item in year_of_release_elem:
    year_of_release.append(item.text)
```

```python
#readgenre
genre=[]
genre_elem=soup.find_all(class_='genre')
for item in genre_elem:
    genre.append(item.text)
```

```python
#read_imdb_rating
imdb_rating=[]
imdb_rating_elem=soup.find_all('div',{'class','inline-block ratings-imdb-rating'})
for item in imdb_rating_elem:
    imdb_rating.append(item.text)
    
```

```python
#read_metascore
metascore=[]
metascore_elem=soup.find_all(class_='inline-block ratings-metascore')
for item in metascore_elem:
    metascore.append(item.text)
    
```

```python
#readsummary
summary=[]
summary_elem=soup.find_all(class_='text-muted')
for item in summary_elem:
    summary.append(item.text)
    
```

```python
#readdirectorandstar
director=[]
director_elem=soup.find_all(class_='lister-item-content')
for item in director_elem:
    director.append(item.text)
```

```python
#read_votes_gross
votes=[]
votes_elem=soup.find_all(class_='sort-num_votes-visible')
for item in votes_elem:
    votes.append(item.text)
votes
```

```python
#creating dataframe
final_array=[]
for Index_No,Movie_Name, Release_Year ,Genre, IMDB_Rating,Metascore, Summary,DirectorandStar,VotesandGross in zip(index_no,movie_name,year_of_release,genre,imdb_rating,metascore,summary,director,votes):
    final_array.append({'index_no':Index_No,'movie_name':Movie_Name,'year_of_release':Release_Year,'genre':Genre,'imdb_rating':IMDB_Rating,'metascore':Metascore,'summary':Summary,'director':DirectorandStar,'votes':VotesandGross})
    
df = pd.DataFrame(final_array)
df
```
